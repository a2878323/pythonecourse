```python
def a(x,*args):
    print('x:',x)
    print('args:',args)
    res = x
    for e in args:
        res = res +e
    return res
        

a(1,2,3,4,5,6)
```

```python
df = data
plt.scatter(df.GrLivArea, df.SalePrice, c = "blue", marker = "s") 
plt.title("Looking for outliers")
plt.xlabel("GrLivArea")
plt.ylabel("SalePrice")
plt.show()
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
colors = le.fit_transform(df.HouseStyle)
colors
```

```python

plt.scatter(df.GrLivArea, df.SalePrice, c = colors, marker = "s") 
plt.title("Looking for outliers")
plt.xlabel("GrLivArea")
plt.ylabel("SalePrice")
plt.show()
```

```python
numeric_cols = df.select_dtypes(exclude=['object']).columns 
fig = plt.figure(figsize = (15,15))
ax = fig.gca()
df[numeric_cols].hist(ax=ax)
fig.tight_layout()
fig.show()
```

```python
categorical_cols = df.select_dtypes(['object']).columns 
fig = plt.figure(figsize=(15,50))
for i, col in enumerate(categorical_cols):
    fig.add_subplot(11,4,1+i)
    df.groupby(col).mean()['SalePrice'].plot.bar(yerr = df.groupby(col).std()) 
fig.tight_layout()
fig.show()
```

```python
import pandas
trainData = pandas.read_csv('train.csv') 
testData = pandas.read_csv('test.csv')
trainData['dataType'] ='train'
testData['dataType'] ='test'
SalePrice = trainData['SalePrice']
fullData = pandas.concat((trainData, testData)).reset_index(drop=True) 
fullData.drop(['SalePrice'], axis=1, inplace=True)
```

```python
see=pd.DataFrame([\
                  fullData.isnull().sum(),\
                  fullData.isnull().sum()/len(fullData)
                 ]
                )
see
```

```python
M=['KitchenQual','Exterior1st','Exterior2nd','MasVnrType']

for col in M:
    print (fullData[col].isnull().sum())
    print (fullData[col].value_counts())
    fullData[col].fillna(fullData[col].value_counts().keys()[0], inplace=True) 
    
```
